## 2. Creacion de Grafico de barras "Precio promedio por tipo de vehículo"

```python
# Diccionario que asigna cada tipo de vehículo con su color correspondiente
colors = {
    'sedan': '#3498db',         # Azul
    'pickup': "#13dd38",        # Verde lima
    'SUV': '#e74c3c',           # Rojo
    'truck': '#f39c12',         # Naranja
    'coupe': '#9b59b6',         # Morado
    'van': "#39d9e4",           # Azul claro
    'convertible': "#bc81f8",   # Rosa
    'hatchback': '#2ecc71',     # Verde
    'wagon': '#1abc9c',         # Turquesa
    'minivan': '#d35400',       # Naranja oscuro
    'other': '#7f8c8d'          # Gris
}

# Calcula el precio promedio por tipo de vehículo
# (eliminar Despues) Contiene: una serie de pandas donde el índice es el tipo de vehículo y el valor es el precio promedio
avg_price_by_type = car_data.groupby('type')['price'].mean().sort_values(ascending=False)

# x_values: lista de tipos de vehículo
x_values = avg_price_by_type.index.tolist()
# y_values: lista de precios promedio por tipo de vehículo
y_values = avg_price_by_type.values.tolist()

# Objeto Figure de Plotly para almacenar el gráfico de barras
Bar_fig = go.Figure()

# Agrega una sola traza de barras con los valores agregados por tipo de vehículo
Bar_fig.add_trace(go.Bar(
    x=x_values,
    y=y_values,
    marker_color=[colors.get(t, "#2aadb6") for t in x_values],
    text=[f"${p:,.0f}" for p in y_values],
    textposition='auto',
    name='Precio promedio'
))

# Actualiza la distribución del gráfico con títulos, estilos y configuración de visualización
Bar_fig.update_layout(title_text='Precio Promedio por Tipo de Vehículo',
                       xaxis_title_text='Tipo de Vehículo', yaxis_title_text='Precio Promedio',
                       yaxis_showgrid=True, yaxis_gridcolor='rgba(200, 200, 200, 0.3)',
                       bargap=0.2
                       )
```
